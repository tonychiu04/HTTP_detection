```python
def fingerprintGeneration(data):
    temp_feature={'host':[],'ip':[],'userAgent':[]}
    for index in range(len(data)):
        for key in fingerprint_keys_index:
            for key_index in fingerprint_keys_index[key]:
                if data[index][key_index] not in temp_feature[key]:
                    temp_feature.setdefault(key,[]).append(data[index][key_index])
    return temp_feature
```
